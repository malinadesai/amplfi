```python
class ProxyOpt:
    def __init__(self, opt_class, **optim_kwargs):
        self.opt_class = opt_class
        self.optim_kwargs = optim_kwargs
    def __call__(self, *args, **kwargs):
        kwargs.update(self.optim_kwargs)
        return self.opt_class(*args, **kwargs)
```

```python
central_pt = 15*sampling_frequency
# we want to add an extra second onto the data window and take the central points
pts_from_center = int((time_duration+1)*sampling_frequency / 2)
windowed_data = []
for ifo in range(len(ifos)):
    newinj_data = []
    psd_list = []
    for inj in range(len(filtered_tsdata[ifo])):
        shortened_data = torch.tensor(filtered_tsdata[ifo][inj][central_pt-pts_from_center:central_pt+pts_from_center], dtype=torch.float64)
        newinj_data.append(shortened_data)
    windowed_data.append(newinj_data)
print('shortened data:', len(windowed_data[0][0]))
num_inj = len(windowed_data[0])
num_inj = int(num_inj)
print('num inj:', num_inj)
```

```python
# creating the psds

psd_list = []
for inj in range(num_inj):
    psd_ifo1 = torch.tensor(filtered_tsdata[0][inj][0:(time_duration+1)*sampling_frequency], dtype = torch.float64)
    psd_ifo2 = torch.tensor(filtered_tsdata[1][inj][0:(time_duration+1)*sampling_frequency], dtype = torch.float64)
    psd_data = torch.stack((psd_ifo1, psd_ifo2)).unsqueeze(0)
    spectral_density = SpectralDensity(sample_rate=sampling_frequency, fftlength=(time_duration+1))
    psd = spectral_density(psd_data)
    psd_list.append(psd)
```

```python
def cast_as_bilby_result(
    samples,
    truth,
    inference_params
):
    injection_parameters = {
            k: float(v) for k, v in zip(inference_params, truth)
        }
    
    priors = {
        param: bilby.core.prior.base.Prior(latex_label=param)
        for param in inference_params
    }
    posterior = dict()
    for idx, k in enumerate(inference_params):
        posterior[k] = samples.T[idx].flatten()
    posterior = pd.DataFrame(posterior)

    r = Result(
        label="PEModel",
        injection_parameters=injection_parameters,
        posterior=posterior,
        search_parameter_keys=inference_params,
        priors=priors,
    )
    r.get_sky_projection(nside=32)
    return r
    
```
